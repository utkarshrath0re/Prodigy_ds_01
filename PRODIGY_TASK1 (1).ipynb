LIFE EXPECTANCY BY REGION (Categorical Variable)

```python
import seaborn as sns
```

```python
my_plot =sns.barplot(data=data1, x='Region', y='Life expectancy at birth (years)')
plt.xticks(rotation=45)
plt.show()
```

Comparison of Birth and Death Rates Across Regions

```python
# Group the data by 'Region' and compare the mean Birth Rate and Death Rate
birth_death_rates_by_region = data1.groupby('Region')[['Birth rate, crude (per 1,000 people)', 'Death rate, crude (per 1,000 people)']].mean().reset_index()

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='value', hue='variable', data=pd.melt(birth_death_rates_by_region, id_vars='Region'), palette='muted')

plt.title('Comparison of Birth and Death Rates Across Regions')
plt.ylabel('Rate (per 1,000 people)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Rate Type', loc='upper right')

plt.show()
```

Comparison of Birth and Death Rates Over the Years

```python
# Group the data by 'Year' and calculate the mean of birth and death rates
birth_death_rates_by_year = data1.groupby('Year')[['Birth rate, crude (per 1,000 people)', 'Death rate, crude (per 1,000 people)']].mean().reset_index()

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='value', hue='variable', data=pd.melt(birth_death_rates_by_year, id_vars='Year'), palette='muted')

plt.title('Comparison of Birth and Death Rates Over the Years')
plt.ylabel('Rate (per 1,000 people)')
plt.xlabel('Year')
plt.legend(title='Rate Type', loc='upper right')

plt.show()
```

Electric Power Consumption Over the Years

```python
# Group the data by 'Year' and calculate the max of Electric power consumption
electric_power_by_year = data1.groupby('Year')['Electric power consumption (kWh per capita)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(electric_power_by_year['Year'], electric_power_by_year['Electric power consumption (kWh per capita)'], marker='o', linestyle='-')

plt.title('Electric Power Consumption Over the Years')
plt.xlabel('Year')
plt.ylabel('Electric Power Consumption (kWh per capita)')
plt.grid(True)

plt.show()
```

 DISTRIBUTION OF LIFE EXPECTANCY 

```python
year = 2012
variable = 'Life expectancy at birth (years)'
data_year = data[data['Year'] == year]
```

```python
#Create A Histogram

plt.figure(figsize=(10, 6))
plt.hist(data_year[variable], bins=20, edgecolor='k', alpha=0.7)
plt.title(f'Distribution of {variable} in {year}')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.show()
```

LIFE EXPECTANCY AT BIRTH 

```python
sns.histplot(data1['Life expectancy at birth (years)'],bins=25,kde=True,color='blue')
plt.title('Distribution of Life Expectancy at Birth')
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('Frequency')
plt.show()
```

Life Expectancy Over the Years

```python
# Group the data by 'Year' and calculate the mean of life expectancy
life_expectancy_by_year = data1.groupby('Year')['Life expectancy at birth (years)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(life_expectancy_by_year['Year'], life_expectancy_by_year['Life expectancy at birth (years)'], marker='o', linestyle='-')

plt.title('Life Expectancy Over the Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.grid(True)

plt.show()
```

GDP (USD) Usage Across Regions

```python
# Group the data by 'Region' and calculate the max GPD
GDP_by_region = data1.groupby('Region')['GDP (USD)'].max().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GDP (USD)', data=GDP_by_region, palette='viridis')
plt.title('GDP (USD) Usage Across Regions')
plt.ylabel('GDP (USD)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

```python
# Group the data by 'Region' and calculate the max GPD per Capita
GDP_percapita_by_region = data1.groupby('Region')['GDP per capita (USD)'].max().reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GDP per capita (USD)', data=GDP_percapita_by_region, palette='viridis')
plt.title('GDP per capita (USD) Usage Across Regions')
plt.ylabel('GDP per capita (USD)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

Income Group Distribution Across Regions

```python
#Group the data by 'Region' and count the incomeGroups
income_group_counts = data1.groupby(['Region', 'IncomeGroup']).size().unstack(fill_value=0)

# Plotting the grouped bar chart
plt.figure(figsize=(14, 8))
income_group_counts.plot(kind='bar', stacked=True, colormap='viridis')

plt.title('Income Group Distribution Across Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Income Group')

plt.show()
```
